```python
import corn
import matplotlib.pyplot as plt
```

```python
# create a new simulator. this has all the variables that you can manipulate
# as defined by the Simulator object in the __init__ function. For example,
# the line below changes volatility to 0.16, but you can change any of the
# appropriate variables.
sim = corn.Simulator(volatility=0.16)

# this fills an array in the simulator with a possible settlement prices.
sim.fill_settlements()

# plots the simulated stock movement with the accumulator being the top line
# and the knockout being the bottom line.
sim.plot_settlements()
```

```python
# we can get a bunch of simulations by running the following for different
# values of n. The result is just the price of the contract over that 
# period of time.
prices = sim.simulate(10000)

# we can create a histogram of prices like this:
theplot = plt.hist(x=prices, bins='auto')
plt.show()
```
